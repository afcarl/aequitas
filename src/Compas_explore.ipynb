```python
import statsmodels.api as sm
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)

df = pd.DataFrame({"v_pred_risk": [0 if a=="Low" else 1 for a in data.v_score_text],
                  "pred_risk": [0 if a=="Low" else 1 for a in data.score_text],
                  "female":  [1 if a=="Female" else 0 for a in data.sex],
                  "over_45":  [1 if a > 45 else 0 for a in data.age],
                  "under_25": [1 if a < 25 else 0 for a in data.age],
                  "race": data.race,
                  "two_year_recid": data.two_year_recid,
                  "priors_count": data['priors_count.1'] ,
                  "misdemeanor": [1 if a=="M" else 0 for a in data.c_charge_degree],
                  "constant": [1 for a in data.id]})
df = pd.get_dummies(df, columns=["race"])
df.describe()
```

```python
#train_cols = [c for c in df.columns if (c !='pred_risk' and c != "race_Caucasian")]
train_cols = ['female',
 'over_45',
 'under_25', 
 'race_African-American',
 'race_Asian',
 'race_Hispanic',
 'race_Native American',
 'race_Other',
 'priors_count',
 'misdemeanor',
 'two_year_recid',
 'constant']
```

```python
logit = sm.Logit(df['pred_risk'], df[train_cols])

# fit the model
result = logit.fit()
result.summary()
```

```python
logit = sm.Logit(df['v_pred_risk'], df[train_cols])

# fit the model
result = logit.fit()
result.summary()
```
